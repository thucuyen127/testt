```python
#Thư viện
import pandas as pd
import seaborn as sns # For Data Visualization
from scipy.stats import norm # for scientific Computing
import matplotlib.pyplot as plt # For Data Visualization
import statsmodels.api as sm
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv('Downloads\data.csv')
data.head()
```

```python
def hist(inp):
    plt.figure()
    ax = sns.distplot(inp)
    plt.axvline(np.mean(inp), color="r", linestyle="dashed", linewidth=5)
    _, max_ = plt.ylim()
    plt.text(
        inp.mean() + inp.mean() / 10,
        max_ - max_ / 10,
        "Mean: {:.2f}".format(inp.mean()),
    )
    return plt.figure
```

```python
#kolmogorov test
from scipy.stats import kstest
 
    
mean1 = data['Screen Size'].mean()
std1 = data['Screen Size'].std()
mean2 = data['Weight'].mean()
std2 = data['Weight'].std()
mean3 = data['Brightness'].mean()
std3 = data['Brightness'].std()
da2 = (data['Weight'] - mean2) /std2
da1 = (data['Screen Size'] - mean1) /std1
da3 = (data['Brightness'] - mean3) /std3

sts1, pv1 = kstest(da1,'norm')
sts2, pv2 = kstest(da2,'norm')
sts3, pv3 = kstest(da3,'norm')

print('statistic of Screen Size = ',sts1)
print('pvalue of Screen Size = ',pv1)
if (pv1 > 0.05):
    print('Không đủ bằng chứng để bác bỏ: Đây là dữ liệu phân phối chuẩn')
else:
    print('Đấy không phải dữ liệu phân phối chuẩn')
print('-'*50)
 
    
print('statistic of Weight = ',sts2)
print('pvalue of Weight = ',pv2)
if (pv2 > 0.05):
    print('Không đủ bằng chứng để bác bỏ: Đây là dữ liệu phân phối chuẩn')
else:
    print('Đấy không phải dữ liệu phân phối chuẩn')
print('-'*50)


print('statistic of Brightness = ',sts3)
print('pvalue ofBrightness = ',pv3)
if (pv3  > 0.05):
    print('Không đủ bằng chứng để bác bỏ: Đây là dữ liệu phân phối chuẩn')
else:
    print('Đấy không phải dữ liệu phân phối chuẩn')
```

```python
#Code kiểm định tính đồng nhất về phương sai của các nhóm

st13, p13 = stats.bartlett(data['Screen Size'], data['Weight'], data['Brightness'])
print('statistic by Bartlett = ',st13)
print('pvalue by Bartlett = ',p13)
if (p13 > 0.05):
    print('Không đủ bằng chứng để bác bỏ: Đây là các dữ liệu đồng nhất')
else:
    print('Đấy không phải là các dữ liệu đồng nhất')
print('-'*50)

st4, p4 = stats.levene(data['Screen Size'], data['Weight'], data['Brightness'])
print('statistic by levene = ',st4)
print('pvalue by levene = ',p4)
if (p4 > 0.05):
    print('Không đủ bằng chứng để bác bỏ: Đây là các dữ liệu đồng nhất')
else:
    print('Đấy không phải là các dữ liệu đồng nhất')
print('-'*50)
```
