```python
#Thư vieenjn và hàm
import pandas as pd
import seaborn as sns # For Data Visualization
from scipy.stats import norm # for scientific Computing
import matplotlib.pyplot as plt # For Data Visualization
import statsmodels.api as sm
import numpy as np
from scipy import stats
from scipy.stats import kstest
from scipy.stats import f_oneway
import statsmodels.api as sm
from statsmodels.formula.api import ols

def hist(inp):
    plt.figure()
    ax = sns.distplot(inp)
    plt.axvline(np.mean(inp), color="r", linestyle="dashed", linewidth=5)
    _, max_ = plt.ylim()
    plt.text(
        inp.mean() + inp.mean() / 10,
        max_ - max_ / 10,
        "Mean: {:.2f}".format(inp.mean()),
    )
    return plt.figure
```

```python
#File owan2
# Biểu đồ
data2 = pd.read_csv('owan02.csv')
print(data2.head())

st1, p1 = stats.shapiro(data2['X1'])
st2, p2 = stats.shapiro(data2['X2'])
st3, p3 = stats.shapiro(data2["X3"])
st4, p4 = stats.shapiro(data2["X4"])
st5, p5 = stats.shapiro(data2["X5"])
```

Xây dựng giả thuyết
$$H_{0}: \text{X1 là dữ liệu phân phối chuẩn}$$
$$H_{a}: \text{X1 không phải là dữ liệu phân phối chuẩn}$$

```python
hist(data2['X1'])
plt.title('Graph of X1')
#qq
da1 = (data2['X1'] - data2['X1'].mean()) /data2['X1'].std()
sm.qqplot(da1, line='45')
plt.show()

print('statistic of X1 = ',st1)
print('pvalue of X1 = ',p1)
if (p1 > 0.05):
    print("Không đủ bằng chứng để bác bỏ H0: X1 là dữ liệu phân phối chuẩn ")
    print('Nên ta có thể phát biểu rằng: X1 là dữ liệu phân phối chuẩn ')
else:
    print('đã có đủ bằng chứng để bác bỏ H0: X1 là dữ liệu phân phối chuẩn')
    print('Đồng nghĩa với việc chúng ta chấp nhận giả thiết Ha')
    print('X1 không phải là dữ liệu phân phối chuẩn')
```
